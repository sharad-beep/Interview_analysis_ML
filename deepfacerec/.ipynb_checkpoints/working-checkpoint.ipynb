```python
# print("hello")
import cv2
img = cv2.imread('facedb/Rob/person2.png') 
plt.imshow(img[:, :, : : -1]) 
```

```python
from deepface import DeepFace
import matplotlib.pyplot as plt
# Analyze the image and get the emotions and their scores.
objs = DeepFace.analyze(img_path="facedb/Rob/person2.png", actions=["age", "gender", "race", "emotion"])

# print(objs)
# Extract the dictionary from the list (assuming there is only one dictionary in the list)
result_dict = objs[0]

# Extract emotions and scores from the dictionary
emotions_and_scores = list(result_dict['emotion'].items())

# Sort the emotions and scores by score, in descending order.
sorted_emotions_and_scores = sorted(emotions_and_scores, key=lambda x: x[1], reverse=True)

# Extract the emotions and scores from the sorted list of tuples.
emotions = [emotion[0] for emotion in sorted_emotions_and_scores]
scores = [emotion[1] for emotion in sorted_emotions_and_scores]

# Print emotions and their corresponding scores
for emotion, score in zip(emotions, scores):
    print(f"Emotion: {emotion}, Score: {score}")


# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(emotions, scores, color='skyblue')
plt.xlabel('Emotions')
plt.ylabel('Scores')
plt.title('Emotions vs. Scores')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Display the bar graph
plt.show()
```
